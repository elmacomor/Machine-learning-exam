```python
import pandas as pd
import numpy as np
import datetime

from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import mean_absolute_error


import matplotlib.pyplot as plt

from sklearn.ensemble import RandomForestClassifier

from sklearn.neighbors import KNeighborsClassifier

from imblearn.over_sampling import SMOTE

import seaborn as sns
```

```python
#1 loading the dataset
df=pd.read_excel('kardiologija_hospitalizacija.xlsx')
```

```python
#checking if the dataset is loaded properly
df.head()
```

```python
#2 analysing dataset size, values and datatypes
df.shape
```

```python
df.info()
```

```python
df=df.drop(columns='dcNyha')
```

```python
#3 data cleanup-->setting null values to mode(object) or mean(float64)
visitedYearMode=df['dmVisitdtYr'].mode()[0]
birthYearMode=df['dmBirthdtYr'].mode()[0]
df['dmVisitdtYr']=df['dmVisitdtYr'].fillna(visitedYearMode)
df['dmBirthdtYr']=df['dmBirthdtYr'].fillna(birthYearMode)
```

```python
df.info()
```

```python
floatColumns=df.select_dtypes('float64').columns
floatColumns
```

```python
df['hsGl'].unique()
```

```python
df['hsGl']=df['hsGl'].replace({datetime.datetime(2017, 3, 5, 0, 0):np.nan})
```

```python
df['hsGl'].unique()
```

```python
df['hsGl']=df['hsGl'].astype(np.float64)
```

```python
floatColumns=df.select_dtypes('float64').columns
floatColumns
```

```python
for col in floatColumns:
    df[col]=df[col].fillna(df[col].mean())
```

```python
df.info()
```

```python
objectColumns=df.select_dtypes('object').columns
objectColumns
```

```python
for col in objectColumns:
    df[col]=df[col].fillna(df[col].mode()[0])
```

```python
df.info()
```

```python
#4 converting categorical variables to numeric
yes_no_cols=[]

for col in objectColumns:
    if set(df[col].unique())=={'Yes', 'No'}:
        yes_no_cols.append(col)
```

```python
yes_no_cols
```

```python
for col in yes_no_cols:
    df[col]=df[col].map({
        'Yes':1,
        'No':0
    })
```

```python
df[yes_no_cols].head()
```

```python
restObjectCols=df[objectColumns].drop(columns=yes_no_cols).columns
restObjectCols
```

```python
featureColumns=df[restObjectCols].drop(columns='hsNyha').columns
featureColumns
```

```python
df=pd.get_dummies(df,columns=featureColumns)
```

```python
df.shape
```

```python
df['hsNyha'].unique()
```

```python
df['hsNyha']=df['hsNyha'].map({
    'NYHA II':0,
    'NYHA III':1,
    'NYHA IV':2
})
```

```python
df['hsNyha'].unique()
```

```python
#5 Splitting the data into training and test sets
X=df.drop(columns=['hsNyha','PatientID'])
Y=df['hsNyha']
```

```python
Xtrain,Xtest,Ytrain,Ytest=train_test_split(X,Y, test_size=0.2, random_state=42)
```

```python
Ytrain.value_counts()
```

```python
sm=SMOTE(random_state=42)
Xtrain_res,Ytrain_res=sm.fit_resample(Xtrain,Ytrain)
```

```python
Ytrain_res.value_counts()
```

```python
#6 Model training using RandomForestClassifier
randomForestModel= RandomForestClassifier(n_estimators=100,random_state=42)
randomForestModel.fit(Xtrain_res,Ytrain_res)
```

```python
#Predictions
predictions=randomForestModel.predict(Xtest)
```

```python
#confusion matrix
matrix=confusion_matrix(Ytest,predictions)
print(matrix)
```

```python
#classification report
print(classification_report(Ytest, predictions))
```

```python
ax=sns.heatmap(matrix, annot=True, cmap='Blues')
```

```python
#accuracy score
accuracy=accuracy_score(Ytest,predictions)
print("Accuracy: ", accuracy)
```

```python
#precision score

precision=precision_score(Ytest,predictions,average='macro')
print("Precission:",precision)
```

```python
#recall score
recall=recall_score(Ytest,predictions,average='macro')
print("Recall:", recall)
```

```python
#f1 score
f1 = f1_score(Ytest, predictions,average='macro')
print("F1-score:", f1)
```
